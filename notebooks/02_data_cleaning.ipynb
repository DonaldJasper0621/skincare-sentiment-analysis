# 02 — Data Cleaning
Removes duplicates, filters low-quality posts, normalises text, and saves a clean CSV.

```python
import pandas as pd
import re

df = pd.read_json('../data/raw/posts_raw.json')
print(f'Raw posts: {len(df)}')
```

```python
# Drop duplicates and deleted/empty posts
df = df.drop_duplicates(subset='post_id')
df = df[df['body'].notna() & (df['body'] != '[deleted]') & (df['body'] != '[removed]')]
df = df[df['body'].str.len() > 50]  # filter very short posts
print(f'After cleaning: {len(df)}')
```

```python
def clean_text(text: str) -> str:
    text = re.sub(r'http\S+', '', text)          # remove URLs
    text = re.sub(r'[^\w\s\-\'%]', ' ', text)  # keep letters, numbers, hyphens
    text = re.sub(r'\s+', ' ', text).strip()
    return text.lower()

df['body_clean'] = df['body'].apply(clean_text)
df.head(3)
```

```python
df.to_csv('../data/processed/posts_clean.csv', index=False)
print('Saved posts_clean.csv ✓')
```
